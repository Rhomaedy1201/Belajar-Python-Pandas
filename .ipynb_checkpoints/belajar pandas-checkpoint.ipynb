# 02 Pemilihan baris (rows selection) pada Data Frame

```python
n_rows = 10
n_cols = 5
cols = tuple('ABCDE')

df = pd.DataFrame(np.random.randint(1, 5, size=(n_rows, n_cols)),
                 columns=cols)

df
```

```python
# mencari nilai pada column A yang nilainya 1 dan 3
df[(df['A'] == 1) | (df['A'] == 3)]
```

```python
# menhilangkan nilai pada column A yang nilainya 1 dan 3
df[~df['A'].isin([1, 3])]
```
